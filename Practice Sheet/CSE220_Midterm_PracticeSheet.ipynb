```python
#task-01

class circularArray:
  def __init__(self,ary,st,size):
    self.ary = ary
    self.st = st
    self.size = size

  def rightshift(self):

    start = self.st
    source = self.ary
    last_idx = (self.size+self.st-1) % len(source)
    k = last_idx 
    for i in range(0,self.size):
      idx = (k + len(source)) % len(source)
      if idx == self.st:
        source[idx] = 0
      else:
        source[idx] = source[idx-1] 
      k -= 1
      start +=1

    print(source)

  def insert(self,elm,new_idx):
    if new_idx < 0 or new_idx > len(self.ary):
      print('Invalid')
    else:
      insert_idx = new_idx +self.st
      last_idx = (self.st+self.size-1) % len(self.ary)
      k = last_idx+1
      for i in range(0,self.size-insert_idx):
        idx  = (k + len(self.ary)) % len(self.ary)
        if idx == insert_idx:
          self.ary[idx] = elm
        else:
          self.ary[idx] = self.ary[idx-1]
        k -= 1

      print(self.ary)
  def rightrotate(self,t):

    for i in range(0,t):
      source = self.ary
      last_idx = (self.size+self.st-1) % len(source)
      k = last_idx+1
      for i in range(0,self.size):
        idx = (k + len(source)) % len(source)
        source[idx] = source[idx-1] 
        k -= 1

      self.st += 1
        
    print(source)

  def leftrotation(self,t):
    for p in range(0,t):
      source = self.ary
      k = -1
      for i in range(0,self.size):
        idx = (self.st + k) % len(self.ary)
        if idx == len(self.ary)-1:
          source[idx] = source[0]
        else:
          source[idx] = source[idx+1]
        k += 1

      self.st -= 1
    print(source)

ary = [25,75,52,25,0,0,46,25,5,19,13,6,7]
c1 = circularArray(ary,6,11)
c2 = circularArray(ary,6,11)
c3 = circularArray(ary,6,11)
c1.rightshift()
c1.insert(62,5)
c2.rightrotate(3)
c3.leftrotation(1)
```

```python
class Node:
  def __init__(self,e,n):
    self.element = e
    self.next = n
```

```python
# task -05

class merge_linedlist:
  def __init__(self,arr1,arr2):

    self.head_1 = Node(None,None)
    current_head_1 = self.head_1
    for i in range(0,len(arr1)):
      node1 = Node(arr1[i],None)
      current_head_1.next = node1
      current_head_1 = current_head_1.next

    self.head_2 = Node(None,None)
    current_head_2 = self.head_2
    for i in range(0,len(arr2)):
      node2 = Node(arr2[i],None)
      current_head_2.next = node2
      current_head_2 = current_head_2.next

  def merge_sort(self):
    current_head_1 = self.head_1.next
    while current_head_1.next != None:
      current_head_1 = current_head_1.next

    current_head_1.next = self.head_2.next

arr1 = [1,2,3,4,5]
arr2 = [6,7,8,9,10]
obj = merge_linedlist(arr1,arr2)
obj.merge_sort()
```

```python
class merge_linkedlist:
  def __init__(self,arr1):

    self.head_1 = Node(arr1[0],None)
    current_head_1 = self.head_1
    for i in range(1,len(arr1)):
      node1 = Node(arr1[i],None)
      current_head_1.next = node1
      current_head_1 = current_head_1.next

  def print_l(self):
    tail = self.head_1
    while(tail.next != None):
      tail = tail.next
    
    print(tail.element)

obj = merge_linkedlist([1,2,3,4,5,6])
obj.print_l()
```
