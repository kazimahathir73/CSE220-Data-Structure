```python
#forwardprint

def forwardprint(source,size,start):

  idx = start
  for i in range(0,size):
    if idx > len(source)-1:
      idx = 0
    print(source[idx],end=',')
    idx+=1

circularArray=[40,50,0,0,0,0,0,0,10,20,30]
forwardprint(circularArray,5,8)
```

```python
#backwardprint

def backwardprint(source,size,start):

  idx = ((start+size)-1) % len(source)     #formula for finding last index of circular array 
  for i in range(0,size):
    if idx < 0:
      idx = len(source)-1
    print(source[idx])
    idx-=1

circularArray=[40,50,60,0,0,0,0,0,10,20,30]
backwardprint(circularArray,6,8)
```

```python
#Resizing

def resize(source,st,size,new_size):
  
  ary = [None]*new_size
  idx_source = st-1
  idx_ary = st-1

  for i in range(0,size):
    
    if idx_source >= len(source)-1:
      idx_source = 0
    else:
      idx_source+=1

    if idx_ary >= len(ary)-1:
      idx_ary = 0
    else:
      idx_ary+=1

    ary[idx_ary] = source[idx_source]
  print(ary)

circ=[20,30,40,50,10] #creating a circular array with start 4 and size 5

resize(circ,4,5,7) 
```

```python
#inserting

def insert(source,st,size,element):
  
  if len(source) != size:
    idx = ((st+size-1)%len(source))
    source[idx] = element

  else:
    ary = [None]*(size+1)
    idx_source = st-1
    idx_ary = st-1

    for i in range(0,size):
      
      if idx_source >= len(source)-1:
        idx_source = 0
      else:
        idx_source+=1

      if idx_ary >= len(ary)-1:
        idx_ary = 0
      else:
        idx_ary+=1

      ary[idx_ary] = source[idx_source]
    ary[((st+size-1)%len(source))] = element
    source = ary

  print(source)

circ=[20,30,40,50,10] 
insert(circ,4,5,60) 
```
