```python
#qes-01
def sumDigit(a,i):
  a = str(a)
  if i >= len(a):
    return 0
  else:
    return int(a[i]) + sumDigit(a,i+1)

print(sumDigit(234,0)) 
```

```python
#qes-01
def sumDigit(a):
  if a <= 0:
    return 0
  else:
    return a%10 + sumDigit(int(a/10))

print(sumDigit(234)) 
```

```python
#ques-02
def bunnyEars(n):
  if n <= 0:
    return 0
  else:
    if n%2 == 0:
      return 3 + bunnyEars(n-1)
    else:
      return 2 + bunnyEars(n-1)

print(bunnyEars(5))
print(bunnyEars(2))
print(bunnyEars(3))
```

```python
#ques-07
def pairStar(a):
  if len(a) <= 1:
    return a[0]
  else:
    if a[0] == a[1]:
      return a[0] + '*' + pairStar(a[1:])
    else:
      return a[0] + pairStar(a[1:])

print(pairStar('xxxxxxgh'))
```

```python
#ques-10
def strCount(a,b):
  if len(a) < len(b):
    return 0
  else:
    if a[0:len(b)] == b:
      return 1 + strCount(a[len(b):],b)
    else:
      return strCount(a[1:],b)

strCount('catcatcatcccch','cat')
```

```python
#ques-14
def check(elem, arr):
  if len(arr) == 0:
    return False
  else:
    if elem == arr[0]:
      return True
    else:
      return check(elem,arr[1:])

def array220(arr):
  if len(arr) == 1:
    return False
  else:
    if check(arr[0]*10,arr[1:]):
      return True
    else:
      return array220(arr[1:])

array220([1, 2, 22])
```

```python
#ques- 23
def stringClean(a):
  if a == '':
    return ''
  else:
    if a[0] in a:
      return stringClean(a[1:])
    else:
      return a[0] + stringClean(a[1:])

print(stringClean("yyzzza"))
```

```python
#ques - 23
def stringClean(string):
  if len(string) == 0:
    return ""
  elif len(string) == 1:
    return string[0] + string[1:]
  else:
    if string[0] == string[1]:
      return string[0] + stringClean(string[2:])
    else:
      return string[0] + stringClean(string[1:])


print(stringClean('hhhheeeellllllo'))
```

```python
# memoization
array = [None]*100

def fibonacci(n):
  if n <= 0:
    return 0
  if n == 1:
    return 1
  if array[n] != None:
    return array[n]
  else:
    array[n] = fibonacci(n-1) + fibonacci(n-2)
    return array[n]

print(fibonacci(99))
    
```
