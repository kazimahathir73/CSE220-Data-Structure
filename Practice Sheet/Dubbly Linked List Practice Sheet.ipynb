```python
class Node:
  def __init__(self, elem, next, prev):
    self.element = elem
    self.next = next
    self.prev = prev
```

```python
#task-01 

class dubblyLinkList1:
  def __init__(self,a):
    self.head = Node(a[0],None,None)

    current_head = self.head
    for i in range(1,len(a)):
      n = Node(a[i],None,None)
      current_head.next = n
      n.prev = current_head 
      current_head = current_head.next
    

  def palindrome(self):
    total = 1
    tail = self.head
    while tail.next != None:
      tail = tail.next
      total+=1

    mid = total//2
    idx = 1
    current_node = self.head
    flag = False
    while current_node != None:
      if idx <= mid:
        if current_node.element == tail.element:
          flag = True
        else:
          flag = False
          break
      current_node = current_node.next
      tail = tail.prev
      idx += 1

    if flag == True:
      print('palindrome')
    else:
      print('Not Palindrome')

ary = [10,20,30,20,10]
n1 = dubblyLinkList1(ary)
n1.palindrome()
```

Given a dummy headed doubly circular linked list, rotate it left by k node (where k is a positive
integer)

```python
class dummydubblyLinkList:
  def __init__(self,a):
    self.head = Node(None,None,None)
    n = Node(a[0],None,None)
    n.prev = self.head
    self.head.next = n
    current_head = self.head.next

    for i in range(1,len(a)):
      n = Node(a[i],None,None)
      current_head.next = n
      n.prev = current_head 
      current_head = current_head.next

    current_head.next = self.head
    self.head.prev = current_head

  def rightshift(self,t):
    for i in range(0,t):
      tail = self.head.prev
      new_tail = tail.prev
      tail.next = self.head.next
      self.head.next.prev = tail
      self.head.next = tail
      tail.prev = self.head
      tail = new_tail
      tail.next = self.head
      self.head.prev = tail

    current_node = self.head.next
    while current_node != self.head:
      print(current_node.element)
      current_node = current_node.next

  def leftshift(self,t):
    for i in range(0,t):
      tail = self.head.prev
      temp_node = self.head.next
      self.head.next = self.head.next.next
      self.head.next.next.prev = self.head
      tail.next = temp_node
      temp_node.prev = tail
      temp_node.next = self.head
      self.head.prev = temp_node
      tail = temp_node

    current_node = self.head.next
    while current_node != self.head:
      print(current_node.element)
      current_node = current_node.next

  def large_node(self):
    max_node = 0
    current_node = self.head.next
    while current_node != self.head:
      if current_node.element > max_node:
        max_node = current_node.element
      current_node = current_node.next

    print(max_node)
```

```python
ary = [10,20,30,40]
n1 = dummydubblyLinkList(ary)
n1.rightshift(3)
```
