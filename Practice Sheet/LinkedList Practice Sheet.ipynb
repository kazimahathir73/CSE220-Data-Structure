**Formula 1-** Connect one node to another **(A with B)** --> **A.next = B**

**Formula 2-** Disconnect one node to another **(A with B)** --> **A.next = None**

**Formula 03-** Three steps for **making new connection with nodes** --->

**1.** Save the connections or nodes to a variable which are connected to the disconnecting node

**2.** Diconnect the node

**3.** Create new connection 

```python
class Node:
  def __init__(self, elem, next):
    self.element = elem
    self.next = next
```

```python
#task-01

class makeNode1:
  def __init__(self,array):

    self.array = array
    self.head = Node(self.array[0],None)

    current_node = self.head
    for i in range(1,len(self.array)):
      obj = Node(self.array[i],None)
      current_node.next = obj
      current_node = current_node.next

  def rightshift(self):
    
    current_node = self.head
    temp_head = self.head 
    tail = None
    while current_node != None:
      if current_node.next == None:
        tail = current_node
      current_node = current_node.next

    self.head = self.head.next
    tail.next = temp_head
    tail.next.next = None

    node2 = self.head
    while node2.next != None:
      print(node2.element,end=',')
      node2 = node2.next
    print(node2.element)

print('<<<<<<<<<<<<<< Task 01 >>>>>>>>>>>>>>>')
array = [10,20,30,40,50,60]
n1 = makeNode1(array)
n1.rightshift()
```

```python
#task-09

class makeNode9:
  def __init__(self,array):

    self.array = array
    self.head = Node(self.array[0],None)

    current_node = self.head
    for i in range(1,len(self.array)):
      obj = Node(self.array[i],None)
      current_node.next = obj
      current_node = current_node.next

  def palindrome(self):
    str1 = ''
    str2 = ''

    total = -1
    current_node = self.head
    while current_node != None:
      total += 1
      current_node = current_node.next

    mid_idx = total // 2  

    count = 0
    current_node1 = self.head
    while current_node1 != None:
      if count <= mid_idx:
        str1 += str(current_node1.element)
      else:
        str2 += str(current_node1.element)
      count += 1
      current_node1 = current_node1.next
    
    if str1 == str2:
      print('Palindrome')
    else:
      print('Not Palindrome')

    

print('<<<<<<<<<<<<<< Task 09 >>>>>>>>>>>>>>>')
ary = [1,2,3,3,2,1]
n9 = makeNode9(ary)
n9.palindrome()
```
