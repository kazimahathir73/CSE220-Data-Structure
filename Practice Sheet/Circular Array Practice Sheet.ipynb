```python
#task-01

def greaterarr(source):
  for idx1 in range(0,len(source)):
    flag_var = True
    for i in range(idx1+1,len(source)+idx1):
      idx2 = i % len(source)
      if source[idx1] < source[idx2]:
        source[idx1] = source[idx2]
        flag_var=False
        break
    if flag_var == True:
      source[idx1]=-1

  return source

arr = [1,2,1]
print(greaterarr(arr))
```

```python
#task-03

def maxsum_diff(source):

  last_sum = 0
  for idx1 in range(0,len(source)):
    temp_sum = 0
    for i in range(idx1,len(source)+idx1):
      idx2 = i % len(source)
      temp_sum += source[idx1]-source[idx2]
      
    if temp_sum > last_sum:
      last_sum = temp_sum

  return last_sum

arr = [10,12,15]
print(maxsum_diff(arr))
```

```python
#task-04

def minmin(source):

  min_diff= 10000
  for idx1 in range(0,len(source)):
    temp_diff = 10000
    for i in range(idx1+1,len(source)+idx1):
      idx2 = i % len(source)
      diff_var = source[idx1] - source[idx2]
      if diff_var < 0:
        diff_var = diff_var*-1
      if diff_var < temp_diff:
        temp_var = diff_var

    if temp_diff < min_diff:
      min_diff = temp_diff

  return min_diff

arr = [10, 12, 13, 15, 10]
print(minmin(arr))
```
