```python
#5
def changePi(s):
  if s=='':
    return ''
  if s[:2]=='pi':
    return '3.14'+changePi(s[2:])
  return s[0]+changePi(s[1:])

changePi("pip")
```

```python
def changePi(a):
  if len(a) <= 1:
    return a[0]
  else:
    if a[0] == 'p' and a[1] == 'i':
      return '3.14' + changePi(a[2:])
    else:
      return a[0] + changePi(a[1:])

changePi('pixxxhspida')
```

```python
#7
def pairStar(s):
  if s=='':
    return ''
  if len(s)>=2:
    if s[0]==s[1]:
      return s[0]+'*'+pairStar(s[1:])
  return s[0]+pairStar(s[1:])
pairStar("aaaa")
```

```python
#9
def countHi2(s):
  if s=='':
    return 0
  if len(s)>=3:
    if s[:3]=='xhi':
      return countHi2(s[3:])
  if len(s)>=2:
    if s[:2]=='hi':
      return 1+countHi2(s[2:])
  return countHi2(s[1:])

countHi2("xhihi")
```

```python
def countHi2(a):
  if len(a) <= 1:
    return 0
  else:
    if a[0] != 'x':
      if a[1] == 'h' and a[2] == 'i':
        return 1 + countHi2(a[2:])
      else:
        return countHi2(a[1:])
    else:
      return countHi2(a[3:])

print(countHi2('axhihi'))
```

```python
#####################15
def endX(s,s1=''):
  if s=='':
    return s1
  if s[0]=='x':
    return endX(s[1:],s1+s[0])
  return s[0]+endX(s[1:],s1+'')
endX("xhixhix")
```
